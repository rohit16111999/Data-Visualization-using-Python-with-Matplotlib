```
import pandas as pd # import module pandas and give it a nickname pd to use in further instructions
```

```
from matplotlib import pyplot as plt # from matplotlib package import pyplot module and name it as plt
```

###**Plot Graph using pyplot module**

```
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y) # plots x in x-axis and y in the y-axis
plt.plot(x, z)
plt.title("Test Plot") # add title to the graph
plt.xlabel("x") # add label in x-coordinate
plt.ylabel("y and z") # add label in y and z-coordinates as y and z are in same coordinate
plt.legend(["This is y", "This is z"]) # tells which represents which data
plt.show() # show the graph
```

###**Data Analysis by using a Larger Dataset**

```
data = pd.read_csv('countries.csv')
```

```
data
```

#### Compare the Population Growth in the USA and China

```
data.country == 'United States'
```

```
us = data[data.country == 'United States']
```

```
china = data[data.country == 'China']
```

```
china
```

```
us
```

```
plt.plot(us.year, us.population / 10**6) # plot data consisting USA data in population per million
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

####To find the Percentage Growth of Each Year Realtive to the First Year

```
us.population
```

```
us.population / us.population.iloc[0] * 100
```

```
plt.plot(us.year, us.population / us.population.iloc[0] * 100) # plot data consisting USA data in population per million
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```

## **Conclusion** : By studying, plotting and visualizing the above data, we can conclude that the population growth of China has increased tremendously in the coming years as compared to the USA in terms of numbers as well as the relative percentage growth from around the year 1962.
